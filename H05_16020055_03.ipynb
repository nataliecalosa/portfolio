```python
#2.Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas
import pandas as pd #import modules
import matplotlib.pyplot as plt
df = pd.read_csv("nilai_uas.csv") #membaca data
c = (df.loc[df["fakultas"] == "FMIPA"]) #mengambil data fakultas FMIPA
c1 = (df.loc[df["fakultas"] == "FTTM"]) #mengambil data fakultas FTTM
c2 = (df.loc[df["fakultas"] == "FSRD"]) #mengambil data fakultas FSRD
c3 = (df.loc[df["fakultas"] == "SAPPK"])#mengambil data fakultas SAPPK
c4 = (df.loc[df["fakultas"] == "SF"])#mengambil data fakultas SF
c5= (df.loc[df["fakultas"] == "STEI"])#mengambil data fakultas STEI
c6 = (df.loc[df["fakultas"] == "FITB"])#mengambil data fakultas FITB
c7 = (df.loc[df["fakultas"] == "FTMD"])#mengambil data fakultas FTMD
c8 = (df.loc[df["fakultas"] == "SBM"])#mengambil data fakultas SBM
c9 = (df.loc[df["fakultas"] == "FTSL"])#mengambil data fakultas FTSL
c10 = (df.loc[df["fakultas"] == "SITHR"])#mengambil data fakultas SITHR
c11 =  (df.loc[df["fakultas"] == "SITHS"])#mengambil data fakultas SITHS
c12 = (df.loc[df["fakultas"] == "FTI"])#mengambil data fakultas FTI
input_data = {}#deklarasi variabel input data
input_data["jumlah"] = [len(c),len(c1),len(c2),len(c3),len(c4),len(c5),len(c6),len(c7),len(c8),len(c9),len(c10),len(c11),len(c12)] #mengisi var input_data dengan array jumlah, dengan array jumlah tersebut berisi jumlah mahasiswa masing-masing fakultas (digunakan fungsi len() untuk mengetahui jumlah data (panjang array))
input_data["fak"]= ["FMIPA","FTTM","FSRD","SAPPK","SF","STEI","FITB","FTMD","SBM","FTSL","SITHR","SITHS","FTI"] #mengisi var input_data dengan array fak dengan urutan yang sesuai dengan array jumlah (urutan fakultas)
df =pd.DataFrame(data=input_data,columns=["jumlah","fak"]) #memasukkan input_data (yang berisi jumlah mahasiswa tiap fak , dan fak) ke dalam dataframe
df.plot(kind="barh", x="fak",y="jumlah",title="Jumlah Mahasiswa Tiap Fakultas") #perintah untuk mengeplot grafik yang akan menampilkan Jumlah Mahasiswa Tiap Fakultas
```

```python
#2.Diagram pie banyaknya mahasiswa masing-masing fakultas
import pandas as pd #import modules
import matplotlib.pyplot as plt
df = pd.read_csv("nilai_uas.csv") #membaca data
c = (df.loc[df["fakultas"] == "FMIPA"]) #mengambil data fakultas FMIPA
c1 = (df.loc[df["fakultas"] == "FTTM"]) #mengambil data fakultas FTTM
c2 = (df.loc[df["fakultas"] == "FSRD"]) #mengambil data fakultas FSRD
c3 = (df.loc[df["fakultas"] == "SAPPK"])#mengambil data fakultas SAPPK
c4 = (df.loc[df["fakultas"] == "SF"])#mengambil data fakultas SF
c5= (df.loc[df["fakultas"] == "STEI"])#mengambil data fakultas STEI
c6 = (df.loc[df["fakultas"] == "FITB"])#mengambil data fakultas FITB
c7 = (df.loc[df["fakultas"] == "FTMD"])#mengambil data fakultas FTMD
c8 = (df.loc[df["fakultas"] == "SBM"])#mengambil data fakultas SBM
c9 = (df.loc[df["fakultas"] == "FTSL"])#mengambil data fakultas FTSL
c10 = (df.loc[df["fakultas"] == "SITHR"])#mengambil data fakultas SITHR
c11 =  (df.loc[df["fakultas"] == "SITHS"])#mengambil data fakultas SITHS
c12 = (df.loc[df["fakultas"] == "FTI"])#mengambil data fakultas FTI
input_data = {}#deklarasi variabel input data
input_data["jumlah"] = [len(c),len(c1),len(c2),len(c3),len(c4),len(c5),len(c6),len(c7),len(c8),len(c9),len(c10),len(c11),len(c12)] #mengisi var input_data dengan array jumlah, dengan array jumlah tersebut berisi jumlah mahasiswa masing-masing fakultas (digunakan fungsi len() untuk mengetahui jumlah data (panjang array))
input_data["fak"]= ["FMIPA","FTTM","FSRD","SAPPK","SF","STEI","FITB","FTMD","SBM","FTSL","SITHR","SITHS","FTI"] #mengisi var input_data dengan array fak dengan urutan yang sesuai dengan array jumlah (urutan fakultas)
df =pd.DataFrame(data=input_data,columns=["jumlah","fak"]) #memasukkan input_data (yang berisi jumlah mahasiswa tiap fak , dan fak) ke dalam dataframe
d = ['green','blue','black','red','yellow','brown','purple','orange','pink','grey','cyan','magenta','black'] #mendeklarasikan array d berisi warna yang ingin digunakan untuk pembagian pie chart
plt.pie(df.jumlah,labels = df.fak,colors = d) #perintah untuk mengeplot df.jumlah menjadi pie chart dengan labels nama fakultas 
```
