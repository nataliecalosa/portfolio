```python
import pandas as pd #import modules
import matplotlib.pyplot as plt
df= pd.read_csv ("covid.csv") #membaca data
```

```python
df["WHO Region"].value_counts().plot(kind = "hist" , title = "Jumlah data per WHO Region", rwidth = 0.8) #perintah mengeplot jumlah WHO Region
```

```python
a = df.loc[(df["Country/Region"] == "China")] #mendeklarasikan data frame baru yang hanya berisi data China
a.plot(kind="line",x="Date", y="Confirmed",title ="Jumlah kasus terkonfirmasi di China") #perintah plot
```

```python
b = df.loc[(df["Date"] == "2020-02-10")]
b["WHO Region"].value_counts().plot(kind="pie",title="Banyaknya data untuk masing-masing WHO Region pada tanggal 10 Febuari 2020")
```

```python
a = df.loc[(df["Country/Region"] == "China")]
a.plot(kind="scatter", x= "Confirmed",y = "New cases",title="Jumlah kasus terkondirmasi terhadap jumlah kasus terkonfirmasi baru di Indonesia")
```
